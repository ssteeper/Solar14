Values of $x_{E1}$ and $x_{E2}$ are chosen such that the arclength on either side of $A$ is the same.
--
The cutting lines shown above represent a plane in 3 dimensions.

In 3 dimensions, the endpoints of the cutting line represent the edges of a cylinder. The cylinder has radius

${\large r = \frac{x_{E2} - x_{E1}}{2}}$
--
and center

$(x,y) = (h,k) = (x_{E1} + r,0)$
--
with arbitrary z.

The semimajor axis is given by half the distance from E1 to E2:

$\large a = \frac{||(E_{1},E_{2})||}{2}$.
--

```python
def ellipse(xE1, xE2):
    return ( hypot(xE2-xE1,f(xE2)-f(xE1))/2 , (xE2-xE1)/2 )
def ellipseArea(a,b):
    return pi*a*b

(a,b) = ellipse(xE1,xE2)
print a, b
print ellipseArea(a,b)
```
