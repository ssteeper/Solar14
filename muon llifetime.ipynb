```python
%pylab inline

from scipy.misc import factorial
import csv
from scipy.optimize import curve_fit

def running_mean(x, N):
    cumsum = numpy.cumsum(numpy.insert(x, 0, 0))
    return (cumsum[N:] - cumsum[:-N]) / N


fname= 'muon_lifetime_data.csv'
fo = open(fname, 'rU')
a = csv.reader(fo)

counts = []
for row in a:
    counts.append(int(row[0]))
```

```python
#raw data

bins = arange(len(counts))
time = 21.36*bins - 66.47
time /= 1000
fig = figure(figsize=(10,8))
scatter(time, counts, s=10)
xlabel('time $\mu$ s')
ylabel('counts')
show()
```

```python
time = time[23:]
counts = counts[23:]
```

```python
#exponential fit

def func(x, a, tau):
    return a * np.exp( -x / tau)

errors = sqrt(counts)
popt, pcov = curve_fit(func, time, counts) # sigma=errors)
a = popt[0]; tau = popt[1]

fig = figure(figsize=(10,8))
scatter(time, counts, s=10)
plot(time, func(time, a, tau), c='red', lw=2)

title('Muon lifetime -- unsmoothed fit', fontsize=18)
xlabel('time ($\\mu$ s)', fontsize=18)
ylabel('counts', fontsize=18)
show()
print round(tau, 3), 'microsec'
```

```python
samples = 10
smoothed_counts = running_mean(counts, samples)

errors = sqrt(smoothed_counts)
popt, pcov = curve_fit(func, time[:-9], smoothed_counts)
a = popt[0]; tau = popt[1]
dtau = pcov[1][1]**.5

fig = figure(figsize=(10,8))
scatter(time[:-9], smoothed_counts, s=10)
plot(time, func(time, a, tau), c='red', lw=2)

title('Muon lifetime -- smoothed fit', fontsize=18)
xlabel('time ($\\mu$ s)', fontsize=18)
ylabel('counts', fontsize=18)
show()
print 'tau = ', round(tau, 3), '+/- ', round(dtau,3), 'microsec'
```

```python
fname= 'random_HIT.csv'
fo = open(fname, 'rU')
a = csv.reader(fo)
random_counts = []
for row in a:
    random_counts.append(int(row[0]))

def poi(x, lam):
    return lam**x * e**(-lam) / factorial(x, exact=False)
x = arange(0,20,.01)

a = hist(random_counts, bins=20, normed=True);close()
b = hist(random_counts, bins=20, normed=False);close()
scale = b[0][0]/a[0][0]
errors = sqrt(b[0])/512

lam =  mean(random_counts)
bins = arange(len(random_counts))
fig = figure(figsize=(12,8))

errorbar(a[1][:-1], a[0], errors, label='data', c='blue', fmt='none')
scatter(a[1][:-1], a[0], c='blue', lw=.5)
plot(x, poi(x, lam), c='red', label='poisson')
scatter(a[1][:-1], poi(a[1][:-1], lam)-a[0] , c='black', label='residuals (poisson - data)')
plot([-5,25], [0,0], '--', c='red')

xlabel('time $\\mu$ s',fontsize=18)
ylabel('counts',fontsize=18)
legend()
xlim(-2,20)
show()
```
