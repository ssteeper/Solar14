```python
#unrotated parabola (90deg incident rays)
#scaled down so distance Pn -> focus = .94m, area of reflector = 1.94 sq m
#at Pn, y'=1, so m=.5, C=0

x = arange(-2,2,.01)
m = .5; C = 0;
y = m*x**2 + C
Ax = 1.0; Ay = .5; Bx = .94; By = .44;
xE1 = .25; yE1 = .032; xE2 = 1.60; yE2= 1.28;
focusx = 0.0; focusy = 0.5;
arclen = integrate.quad(lambda x: sqrt(1+(2*m*x)**2), xE1, xE2)[0]
arclen1 = integrate.quad(lambda x: sqrt(1+(2*m*x)**2), xE1, Ax)[0]
arclen2 = integrate.quad(lambda x: sqrt(1+(2*m*x)**2), Ax, xE2)[0]
print 'eq arc length e1->B: %.3f' % arclen1
print 'eq arc length B->e2: %.3f' % arclen2

#plotting
fig = plt.figure(figsize=(16, 10), dpi=100)
plot(x,y,label = '$y=.5x^2$')  #parabola
plot(array(Ax),array(Ay), '+', c='red', ms=10,lw = 5, label = '$P_n (1.0, .50)$')  #Pn
plot([xE1, xE2], [yE1, yE2], color='black', label='$y=.94x-.20$')    #cutting line
plot(array(focusx),array(focusy), 'd', c='red')  #focus
plot(array(Bx),array(By), '*', c='green', label = '$B (.94, .44)$')  #pt B

annotate('$F (0.0,0.5)$', (0, .42), fontsize=16)
annotate('North', (.1,1.3), fontsize=16)
annotate('$y=.5x^2$', (1.32, .80), fontsize=16)
annotate('$y=.94x-.20$', (.67, .84), fontsize=16)
annotate('$B (.94, .44)$', (.95, .36), fontsize=16)
annotate('$A (1.0, .50)$', (1.05,.48), fontsize=16)
annotate('$E1 (.25, 0.03)$', (.25,-0.05), fontsize=16)
annotate('$E2 (1.60, 1.28)$', (1.62,1.24), fontsize=16)
annotate('parabolic arc length: %.3f' % arclen, (1.2, 0.0))
annotate('E1->A: %.3f' % arclen1, (1.2, -0.1))
annotate('A->E2: %.3f' % arclen2, (1.2, -0.2))
arrow(.25, 1.25, -.25, 0)
xlim(-.25, 2.0)
ylim(-.25, 1.5)
show()
```

```python
xE1s = arange(0,Bx,.001)  #adjust if slow
xE2s = arange(Bx,1.75, .001)  #adjust if slow

#---SUMMER---
alpha = (maxsd) * degtorad 
rotMatrix = array([[cos(alpha), -sin(alpha)],
        [ sin(alpha), cos(alpha)]], np.float32)

#rotate A
newA = array([Ax, Ay-focusy]).dot(rotMatrix)
AXsum = newA[0]; AYsum = newA[1]+focusy;
angleatA = (45 - maxsd/2) * degtorad
yprime = tan(angleatA)
#determine new eqn for parabola
msum = yprime/(2*AXsum)
Csum = AYsum - msum*AXsum**2
Ysum = msum*x**2 + Csum

narclen1 = []; narclen2=[];
#find endpoints
for nxe1 in xE1s:
    narclen1.append(integrate.quad(lambda x: sqrt(1+(2*msum*x)**2), nxe1, AXsum)[0])
for nxe2 in xE2s:
    narclen2.append(integrate.quad(lambda x: sqrt(1+(2*msum*x)**2), AXsum, nxe2)[0])
        
sarclen1 = [];sarclen2 = [];
for trial in narclen1:
    sarclen1.append(trial-arclen1)
for trial in narclen2:
    sarclen2.append(trial-arclen2)

minarclen1 = min(sarclen1, key=abs) #gets value closest to 0 (so this is the arclen with least error)
minarclen2 = min(sarclen2, key=abs)
ndx1 = sarclen1.index(minarclen1)  #get index of best e1
ndx2 = sarclen2.index(minarclen2)
XE1sum = xE1s[ndx1]; XE2sum = xE2s[ndx2]
YE1sum = msum*XE1sum**2 + Csum; YE2sum = msum*XE2sum**2 + Csum

arclensum1 = integrate.quad(lambda x: sqrt(1+(2*msum*x)**2), XE1sum, AXsum)[0]
arclensum2 = integrate.quad(lambda x: sqrt(1+(2*msum*x)**2), AXsum, XE2sum)[0]

#---WINTER---
alpha = (-maxsd) * degtorad 
rotMatrix = array([[cos(alpha), -sin(alpha)],
        [ sin(alpha), cos(alpha)]], np.float32)

#rotate A
newA = array([Ax, Ay-focusy]).dot(rotMatrix)
AXwin = newA[0]; AYwin = newA[1]+focusy;
angleatA = (45 + maxsd/2) * degtorad
yprime = tan(angleatA)
#determine new eqn for parabola
mwin = yprime/(2*AXwin)
Cwin = AYwin - mwin*AXwin**2
Ywin = mwin*x**2 + Cwin

narclen1 = []; narclen2=[];
#find endpoints
for nxe1 in xE1s:
    narclen1.append(integrate.quad(lambda x: sqrt(1+(2*mwin*x)**2), nxe1, AXwin)[0])
for nxe2 in xE2s:
    narclen2.append(integrate.quad(lambda x: sqrt(1+(2*mwin*x)**2), AXwin, nxe2)[0])
        
sarclen1 = [];sarclen2 = [];
for trial in narclen1:
    sarclen1.append(trial-arclen1)
for trial in narclen2:
    sarclen2.append(trial-arclen2)

minarclen1 = min(sarclen1, key=abs) #gets value closest to 0 (so this is the arclen with least error)
minarclen2 = min(sarclen2, key=abs)
ndx1 = sarclen1.index(minarclen1)  #get index of best e1
ndx2 = sarclen2.index(minarclen2)
XE1win = xE1s[ndx1]; XE2win = xE2s[ndx2]
YE1win = mwin*XE1win**2 + Cwin; YE2win = mwin*XE2win**2 + Cwin

arclenwin1 = integrate.quad(lambda x: sqrt(1+(2*mwin*x)**2), XE1win, AXwin)[0]
arclenwin2 = integrate.quad(lambda x: sqrt(1+(2*mwin*x)**2), AXwin, XE2win)[0]
```
