```python
#fname = 'SENSORS data/DATA_shower_nofan.txt'
#fname = 'SENSORS data/DATA_bandpractice2.txt'
#fname = 'SENSORS data/DATA_after_shower.txt'
fname = 'SENSORS data/slo op and jordans.txt'

reader = csv.reader(open(fname, 'rU'))

a=[]; tmax = 0;
i=0
for row in reader:
    if len(row) is 6 and len(row[0]) > 0 and i != 0:
        if float(row[0]) > tmax:
            a.append(row)
            tmax = float(row[0])
    i+=1
a = array(a)
```
